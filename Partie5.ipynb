```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('supermarket_sales.csv')
data.head()
```

```python
sns.scatterplot(data=data, x='gross income', y='Total')
plt.title('Relation entre le revenue et le total')
plt.grid()
```

```python
sns.scatterplot(data=data, x='Quantity', y='Total')
plt.title('Relation entre la quantite et le totale')
plt.grid()
```

```python
sns.scatterplot(data=data, x='Rating', y='Total')
plt.title('Relation entre la quantite et la note')
plt.grid()
```
