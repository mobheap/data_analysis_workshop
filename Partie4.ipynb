```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('supermarket_sales.csv')
df.columns
```

```python
sns.countplot(data=df, x='City')
plt.title('Nombre de transactions vs City')
plt.xlabel('Ville')
plt.ylabel('Nombre de transactions')
plt.show()
```

```python
sns.countplot(data=df, x='Branch')
plt.title('Nombre de transactions vs Branch')
plt.xlabel('Branche')
plt.ylabel('Nombre de transactions')
plt.show()
```

```python
plt.figure(figsize=(16,5))
sns.countplot(data=df, x='Product line')
plt.title('Nombre de transactions vs Product Line')
plt.xlabel('Ligne de produit')
plt.ylabel('Nombre de transactions')
plt.show()
```

```python
sns.countplot(data=df, x='Customer type')
plt.title('Nombre de transactions vs Customer Type')
plt.xlabel('Type de client')
plt.ylabel('Nombre de transactions')
plt.show()
```

```python
sns.barplot(data=df, x='City', y='Total', ci=None)
plt.title('Total des transactions vs City')
plt.xlabel('Ville')
plt.ylabel('Total des transactions')
plt.show()
```

```python
sns.barplot(data=df, x='Branch', y='Total', ci=None)
plt.title('Total des transactions vs Branch')
plt.xlabel('Branche')
plt.ylabel('Total des transactions')
plt.show()
```

```python
plt.figure(figsize=(16,5))
sns.barplot(data=df, x='Product line', y='Total', ci=None)
plt.title('Total des transactions vs Product Line')
plt.xlabel('Ligne de produit')
plt.ylabel('Total des transactions')
plt.show()
```

```python
sns.barplot(data=df, x='Customer type', y='Total', ci=None)
plt.title('Total des transactions vs Customer Type')
plt.xlabel('Type de client')
plt.ylabel('Total des transactions')
plt.show()
```
