```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('supermarket_sales.csv')
data.head()
```

```python
plt.figure(figsize=(10,4))
```

```python
sns.scatterplot(data=data, x='Tax 5%', y='Total')
plt.xlabel('Tax')
plt.ylabel('Total')
plt.title('Relation entre le totale et les taxes')
plt.grid()
```

Oui, il semble y avoir une corrélation entre les deux variables

La corrélation observée est positive, ce qui signifie qu'à mesure que la valeur des taxes augmente, le montant total semble également augmenter.

L'interprétation de cette corrélation positive indique que, en règle générale, lorsque le montant total augmentent,  les taxes à 5 % augmente également. Cela peut refléter une situation où les taxes sont calculées sur une base de pourcentage du total

```python
sns.histplot(data=data, x='Total')
plt.xlabel('Total')
plt.ylabel('Frequence')
plt.title('Distribution des Totales')
plt.grid()
plt.show()
```

L'histogramme montre une distribution asymétrique de la variable "Total", avec une concentration importante des montants entre 0 et 200. 

Les factures semblent être principalement concentrées sur des montants faibles (compris entre 0 et 200), ce qui suggère que la majorité des transactions sont de petits montants.
